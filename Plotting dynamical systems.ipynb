```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import ma

#Vector field----------------------------------------------------------------------------------
n,k=5,.2
X, Y = np.meshgrid(np.arange(-n, n, k), np.arange(-n, n, k))
U = X+Y-2*X**2
V = -2*X+Y+3*Y**2
name='punto1'

title="Sistema: $x'=x+y-2x^2, y'=-2x+y+3y^2$"

plt.figure(figsize=(12,6))

plt.title(title)
Q = plt.quiver(X[::3, ::3], Y[::3, ::3], U[::3, ::3], V[::3, ::3],color='b' 
             , headwidth=3, headlength=5 ,pivot='mid', units='inches',alpha=0.4, label='Vector field')
#qk = plt.quiverkey(Q, 0.9, 0.9, 1, r'$1 \frac{m}{s}$', labelpos='E',
#                   coordinates='figure')
#plt.scatter(X[::3, ::3], Y[::3, ::3], color='r', s=2)
#Nullclinas-------------------------------------------------------------------------------------
w=[(-n*t/100) + n*(1-t/100) for t in range(100)]
y1 = [2*x**2 -x for x in w]
x1 = [(y+3*y**2)/2 for y in w]

plt.plot(w,y1,'r',alpha=0.8,label="$x'=0$")
plt.plot(x1,w,'g',alpha=0.8,label="$y'=0$")
#Eq points--------------------------------------------------------------------------------------
plt.scatter(0,0,color='r',s=40)
plt.scatter(0.85472,0.606374,color='r',s=40,label='Puntos Eq.')

plt.xlim(-n, n)
plt.ylim(-n, n)

plt.xlabel('$x$'), plt.ylabel('$y$')
plt.legend(loc=3)
name=name+'.pdf'
plt.savefig(name, format='pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
#Vector field----------------------------------------------------------------------------------
n,k=5,.2
X, Y = np.meshgrid(np.arange(-n, n, k), np.arange(-n, n, k))
U = Y**2
V = -(2/3)*X
name='punto2'

title="Sistema: $x'=y^2, y'=-\\frac{2}{3}x$"

plt.figure(figsize=(12,6))

plt.title(title)
Q = plt.quiver(X[::3, ::3], Y[::3, ::3], U[::3, ::3], V[::3, ::3],color='b' 
             , headwidth=3, headlength=5 ,pivot='mid', units='inches',alpha=0.4, label='Vector field')
#qk = plt.quiverkey(Q, 0.9, 0.9, 1, r'$1 \frac{m}{s}$', labelpos='E',
#                   coordinates='figure')
#plt.scatter(X[::3, ::3], Y[::3, ::3], color='r', s=2)
#Nullclinas-------------------------------------------------------------------------------------
w=[(-n*t/100) + n*(1-t/100) for t in range(1000)]
w_=[w[t] for t in range(0,100,5)]
y1 = [0 for x in w]
x1 = [0 for y in w]

plt.plot(w,y1,'y',alpha=1,label="$x'=0$")
plt.plot(x1,w,'g',alpha=0.8,label="$y'=0$")
#Eq points--------------------------------------------------------------------------------------
plt.scatter(0,0,color='r',alpha=0.8,s=40,label='Puntos Eq.')
#Solutions--------------------------------------------------------------------------------------
K=[-7, 0, 7]
for k in K:
    if k==0:
        plt.plot(w,[(k-x**2)**(1/3) for x in w],'r',alpha=0.8,label='$y^3+x^2=k$, $k$=-7,0,7')
        plt.plot(w_,[(k-x**2)**(1/3) for x in w_],'r>',alpha=0.8)
    else:
        plt.plot(w,[(k-x**2)**(1/3) for x in w],'r',alpha=0.8)
        plt.plot(w_,[(k-x**2)**(1/3) for x in w_],'r>',alpha=0.8)
    plt.legend(loc=3)

plt.xlim(-n, n)
plt.ylim(-n, n)

plt.xlabel('$x$'), plt.ylabel('$y$')
plt.legend()
name=name+'.pdf'
plt.savefig(name, format='pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
#Vector field----------------------------------------------------------------------------------
n,k=3.5,.2
X, Y = np.meshgrid(np.arange(-n, n, k), np.arange(-n, n, k))
U = X**2+Y**2-4
V = Y-2*X
name='punto3'

title="Sistema: $x'=x^2+y^2-4, y'=y-2x$"

plt.figure(figsize=(9,9))

plt.title(title)
Q = plt.quiver(X[::3, ::3], Y[::3, ::3], U[::3, ::3], V[::3, ::3],color='b' 
             , headwidth=3, headlength=5 ,pivot='mid', units='inches',alpha=0.4, label='Vector field')
#qk = plt.quiverkey(Q, 0.9, 0.9, 1, r'$1 \frac{m}{s}$', labelpos='E',
#                   coordinates='figure')
#plt.scatter(X[::3, ::3], Y[::3, ::3], color='r', s=2)
#Nullclinas-------------------------------------------------------------------------------------
w=[(-2*t/100) + 2*(1-t/100) for t in range(1000)]
y1 = [2*x for x in w]
x1 = [np.sqrt(4-y**2) for y in w]
x2 = [-np.sqrt(4-y**2) for y in w]

plt.plot(w,y1,'r',alpha=0.8,label="$y'=0$")
plt.plot(w,x1,'g',alpha=0.8,label="$x'=0$")
plt.plot(w,x2,'g',alpha=0.8)
#Eq points--------------------------------------------------------------------------------------
plt.scatter(-0.894427,-1.78885,color='r',s=40)
plt.scatter(0.894427,1.78885,color='r',s=40,label='Puntos Eq.')

plt.xlim(-n, n)
plt.ylim(-n, n)

plt.xlabel('$x$'), plt.ylabel('$y$')
plt.legend(loc=3)
name=name+'.pdf'
plt.savefig(name, format='pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 8
X, Y = np.mgrid[0:n, 0:n]
T = np.arctan2(Y - n / 2., X - n/2.)
R = 10 + np.sqrt((Y - n / 2.0) ** 2 + (X - n / 2.0) ** 2)
U, V = R * np.cos(T), R * np.sin(T)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.quiver(X, Y, U, V, R, alpha=.5)
plt.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=.)

plt.xlim(-1, n)
plt.xticks(())
plt.ylim(-1, n)
plt.yticks(())

plt.show()
```
